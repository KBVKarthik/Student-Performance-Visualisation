```python
#Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading the dataset
data = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
```

```python
#Previewing the dataset
data.head()
```

```python
sns.displot(data = data, x = 'math score', kde = False, bins = 10)
plt.xlabel('Math Scores')
plt.ylabel('Count')
plt.title('Distribution of Math scores')
plt.show()
```

```python
sns.displot(data = data, x = 'reading score', kde = False, bins = 5)
plt.xlabel('Reading Scores')
plt.ylabel('Count')
plt.title('Distribution of Reading scores')
plt.show()
```

```python
sns.displot(data = data, x = 'writing score', kde = False, bins = 5)
plt.xlabel('Writing score')
plt.ylabel('Count')
plt.title('Distribution of Writing scores')
plt.show()
```

```python
sns.barplot(data = data, x='gender',y='math score')
plt.show()
```

```python
sns.barplot(data = data, x='parental level of education',y='math score')
plt.xlabel('Parental level of Education')
plt.ylabel('Math Scores')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.boxplot(data = data,x='race/ethnicity',y='reading score')
plt.show()
```

```python
sns.violinplot(x='test preparation course', y='writing score', hue='gender', data=data, split=True)
plt.title('Writing Scores by Test Preparation Course and Gender')
plt.xlabel('Test Preparation Course')
plt.ylabel('Writing scores')
plt.show()
```

```python
sns.pairplot(data = data[['math score','reading score','writing score']])
plt.show()
```
